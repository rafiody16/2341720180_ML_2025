####**Deskripsi Tugas**
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

- Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

- Lakukan proses encoding pada kolom "diagnosis".

- Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

####Load Data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('wbc.csv')
df.head()
```

```python
df.info()
```

####Memisahkan variabel yang dapat digunakan dengan variabel yang tidak dapat digunakan

Di sini kita akan semua variabel kecuali variabel id dan unnamed: 32. Karena variabel id tidak memiliki korelasi dengan diagnosis kanker. Sedangkan variabel unnamed: 32 merupakan variabel yang tidak memiliki data.

```python
df = df.drop(columns=['id', 'Unnamed: 32'])

df.head()
```

####Encoding variabel Diagnosis

```python
le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

Setelah dilakukan encoding, nilai pada variabel diagnosis yang semula adalah Malignant (M) berubah menjadi 1 dan Benign (B) menjadi 0.

####Standarisasi

```python
# Memisahkan variabel diagnosis agar tidak ikut standarisasi
x = df.drop(columns=['diagnosis'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

X_scaled = pd.DataFrame(X_scaled, columns=x.columns)

X_scaled.head()
```
